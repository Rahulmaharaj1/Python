Loops in Python

Loops are used to repeat a block of code multiple times without rewriting it. They make your programs efficient and easy to maintain.

Why We Need Loops

Imagine you want to print numbers 1 to 5. Without a loop:

print(1)
print(2)
print(3)
print(4)
print(5)

✅ Works, but repetitive.

With a loop:

for i in range(1, 6):
    print(i)

Much shorter, easier to change, and works for bigger tasks automatically.

1. While Loop

Repeats a block of code as long as a condition is True.

Checks the condition before each iteration.

Syntax:

while condition:
    # code to repeat

Example – Print 1 to 5:

i = 1
while i <= 5:
    print(i)
    i += 1  # important to avoid infinite loop

Use case: When you don’t know in advance how many times you need to repeat something.

Example – Ask user until they type 'exit':

command = ''
while command != 'exit':
    command = input("Enter a command (type 'exit' to stop): ")
    print("You entered:", command)
2. For Loop

Repeats a block of code a fixed number of times.

Iterates over a sequence: numbers, list, string, tuple, etc.

Syntax:

for variable in sequence:
    # code to repeat

Example 1 – Using range():

for i in range(1, 6):
    print(i)

Example 2 – Iterating over a list:

fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)

# While loop example -> program to print the table
# Program -> Sum of all digits of a given number
# Program -> keep accepting numbers from users till he/she enters a 0 and then find the avg

```python
# Take input from the user and convert it to an integer
number = int(input('Enter the number: '))

# Initialize a counter variable
i = 1

# While loop runs while i is less than 11 (1 to 10)
while i < 11:
    # Print the multiplication in the format: number * i = result
    print(number, '*', i, '=', number * i)
    
    # Increment the counter to move to the next multiplication
    i += 1
```

```python
# Initialize counter
x = 1

# While loop runs as long as x is less than 3
while x < 3:
    print(x)   # Print the current value of x
    x += 1     # Increment x

# Else block runs **after the while loop finishes normally**
else:
    print('Limit crossed')   # This runs when the condition x < 3 becomes False
```

```python
# Import random module to generate a random number
import random

# Generate a random integer between 1 and 100
jackpot = random.randint(1, 100)

# Take the user's first guess
guess = int(input('Guess karo (1-100): '))

# Initialize a counter to track attempts
counter = 1

# Loop runs until the user's guess matches the jackpot number
while guess != jackpot:
    if guess < jackpot:
        print('Galat! Guess higher')  # Tell user to guess higher
    else:
        print('Galat! Guess lower')   # Tell user to guess lower

    # Ask for the next guess
    guess = int(input('Guess karo: '))
    
    # Increment attempt counter
    counter += 1

# Else block runs when the user guesses correctly
else:
    print('Correct guess! 🎉')
    print('Attempts:', counter)
```

Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

```python
# Initial population
curr_pop = 10000

# Loop from 10 down to 1 (10 years), step -1
for i in range(10, 0, -1):
    # Print the current year (i) and population
    print(i, curr_pop)
    
    # Reduce population by 10%
    curr_pop = curr_pop - 0.1 * curr_pop
    # Equivalent to: curr_pop *= 0.9
```
