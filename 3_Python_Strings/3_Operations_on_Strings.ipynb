Operations on Strings
---------------------->

Arithmetic Operations: Strings support concatenation (+) and repetition (*).

Relational Operations: Strings can be compared using <, >, <=, >=, ==, !=.

Logical Operations: Strings can be used in logical expressions with and, or, not.

Loops on Strings: Strings can be iterated character by character using for or while loops.

Membership Operations: Use in or not in to check if a substring exists within a string.

```python
print('delhi' + ' ' + 'mumbai')
```

```python
print('delhi'*5)
```

```python
print("*"*50)
```

```python
'delhi' != 'delhi'
```

```python
'mumbai' > 'pune'
# lexiographically
```

```python
'Pune' > 'pune'
```

```python
'hello' and 'world'
```

```python
'hello' or 'world'
```

```python
'' and 'world'
```

```python
'' or 'world'
```

```python
'hello' or 'world'
```

```python
not 'hello'
```

```python
for i in 'hello':
  print(i)
```
