count() – Returns the number of times a substring appears in the string.

find() – Returns the lowest index of a substring; returns -1 if not found.

index() – Returns the lowest index of a substring; raises an error if not found.

```python
s = 'hello world'

# count() → number of times a substring appears
print(s.count('l'))   # Output: 3

# find() → lowest index of substring, returns -1 if not found
print(s.find('o'))    # Output: 4
print(s.find('z'))    # Output: -1

# index() → lowest index of substring, raises error if not found
print(s.index('w'))   # Output: 6
# print(s.index('z'))  # ❌ This would raise ValueError
```

Explanation

startswith(substring) → Returns True if the string begins with the given substring.

endswith(substring) → Returns True if the string ends with the given substring.

```python
s = "hello world"

# Check if string starts with 'hello'
print(s.startswith('hello'))   # Output: True

# Check if string ends with 'world'
print(s.endswith('world'))     # Output: True

# Check with a substring not matching
print(s.startswith('world'))   # Output: False
print(s.endswith('hello'))     # Output: False
```

Explanation

format() → Inserts values into placeholders {} in a string.

Allows positional, keyword, or indexed replacement.

Useful for creating dynamic and readable strings.

```python
# Positional formatting
name = "Alice"
age = 25
print("My name is {} and I am {} years old.".format(name, age))

# Keyword formatting
print("My name is {n} and I am {a} years old.".format(n=name, a=age))

# Indexed formatting
print("Coordinates: {0}, {1}".format(10, 20))
```

Explanation

isalnum() → Returns True if all characters are letters or numbers (no spaces or symbols).

isalpha() → Returns True if all characters are letters only.

isdigit() → Returns True if all characters are digits only.

isidentifier() → Returns True if the string is a valid Python identifier (can be used as a variable

```python
s1 = "Hello123"
s2 = "Hello"
s3 = "12345"
s4 = "my_var"
s5 = "123abc!"

print(s1.isalnum())       # True (letters + numbers)
print(s2.isalpha())       # True (letters only)
print(s3.isdigit())       # True (digits only)
print(s4.isidentifier())  # True (valid Python variable name)
print(s5.isalnum())       # False (contains symbol '!')
```

Explanation

split(separator) → Splits a string into a list of substrings using the given separator (default is whitespace).

join(iterable) → Joins elements of an iterable (like list or tuple) into a single string with the string as a separator.

```python
s = "hello world python"

# Split the string into a list of words
words = s.split()  # default separator is space
print(words)       # Output: ['hello', 'world', 'python']

# Join the list of words into a string with hyphens
joined = "-".join(words)
print(joined)      # Output: hello-world-python
```

Explanation

replace(old, new[, count]) → Returns a new string where occurrences of old are replaced with new.

Optional count limits the number of replacements.

Strings are immutable, so replace() does not modify the original string

```python
s = "hello world python world"

# Replace all occurrences of 'world' with 'universe'
new_s = s.replace('world', 'universe')
print(new_s)  # Output: hello universe python universe

# Replace only the first occurrence
new_s2 = s.replace('world', 'universe', 1)
print(new_s2) # Output: hello universe python world
```

Explanation

strip() → Removes leading and trailing whitespace (spaces, tabs, newlines) from a string.

Can also remove specific characters by passing them as an argument.

Does not modify the original string; it returns a new string.

```python
s = "   hello world   "

# Remove leading and trailing whitespace
clean_s = s.strip()
print(clean_s)       # Output: 'hello world'

# Remove specific characters
s2 = "##hello##"
clean_s2 = s2.strip('#')
print(clean_s2)      # Output: 'hello'
```

Example Programs

```python
# Find the length of a given string without using the len() function


# Take input from user
s = input('Enter the string: ')

# Initialize a counter
counter = 0

# Loop through each character in the string
for i in s:
    counter += 1  # Increment counter for each character

# Print the total length
print('Length of string is', counter)
```

```python
# extracts the username part of an email (everything before the @).


# Take email input from the user
s = input('Enter the email: ')

# Find the position of '@'
pos = s.index('@')  # index() gives the position of first occurrence of '@'

# Slice the string from start up to '@'
username = s[0:pos]

# Print the username
print(username)
```

```python
# counts how many times a specific character appears in a string

# Take input string and the character to search
s = input('Enter the string: ')
term = input('Which character would you like to search for? ')

# Initialize counter
counter = 0

# Loop through each character in the string
for i in s:
    if i == term:
        counter += 1  # Increment counter if character matches

# Print the frequency
print('Frequency of', term, 'is', counter)
```

```python
# removes a specific character from a string.

# Take input string and character to remove
s = input('Enter the string: ')
term = input('Which character would you like to remove? ')

# Initialize an empty result string
result = ''

# Loop through each character in the string
for i in s:
    if i != term:       # Only add characters that are not the term
        result += i

# Print the final string
print(result)
```

```python
# hecks whether a string is a palindrome (reads the same forwards and backwards)

# Take input string
s = input('Enter the string: ')

# Initialize flag as True
flag = True

# Loop through first half of the string
for i in range(0, len(s)//2):
    if s[i] != s[len(s) - i - 1]:  # Compare character from start and end
        flag = False
        print('Not a Palindrome')
        break

# If no mismatch was found, it is a palindrome
if flag:
    print('Palindrome')
```

```python
# This code counts words without using split() by manually building words. Let me explain clearly:

# Take input string
s = input('Enter the string: ')

# Initialize an empty list to store words and a temporary string
L = []
temp = ''

# Loop through each character
for i in s:
    if i != ' ':       # If character is not a space, add to temp
        temp += i
    else:              # If space, word ended
        if temp != '': # Avoid adding empty strings for multiple spaces
            L.append(temp)
        temp = ''      # Reset temp for next word

# Append the last word after the loop
if temp != '':
    L.append(temp)

# Print the list of words
print(L)
print('Number of words:', len(L))
```

```python
# converts a string to title case (first letter of each word capitalized) without using title().

# Take input string
s = input('Enter the string: ')

# Initialize a list to store processed words
L = []

# Loop through each word in the string
for i in s.split():                     # Split string into words
    L.append(i[0].upper() + i[1:].lower())  # Capitalize first letter, make rest lowercase

# Join the words back into a single string
print(" ".join(L))
```

```python
# converts an integer to a string manually without using str()

# Take integer input
number = int(input('Enter the number: '))

# String containing digits
digits = '0123456789'

# Initialize an empty result string
result = ''

# Loop until number becomes 0
while number != 0:
    result = digits[number % 10] + result  # Get last digit and add to result
    number = number // 10                   # Remove last digit from number

# Print the resulting string and its type
print(result)        # Output: number as string
print(type(result))  # Output: <class 'str'>
```
