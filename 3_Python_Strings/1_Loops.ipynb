Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

```python
# Initial population
curr_pop = 10000

# Loop from 10 down to 1 (10 years), stepping down by 1
for i in range(10, 0, -1):
    # Print the current year and the current population (rounded to 2 decimals)
    print(i, round(curr_pop, 2))
    
    # Decrease population by dividing by 1.1 (approx 9.09% decrease)
    curr_pop = curr_pop / 1.1
```

Sequence sum
1/1! + 2/2! + 3/3! + ...

```python
# Take input from user
n = int(input('Enter n: '))

# Initialize result to accumulate the sum
result = 0

# Initialize factorial as 1 (0! = 1)
fact = 1

# Loop from 1 to n (inclusive)
for i in range(1, n + 1):
    fact = fact * i         # Calculate factorial i!
    result = result + i / fact  # Add term i / i! to result

# Print the final result
print(result)
```

Pattern 1
***
****
***

```python
# Take input for number of rows
rows = int(input('Enter number of rows: '))

# Outer loop controls the number of rows
for i in range(1, rows + 1):
    # Inner loop prints stars equal to the current row number
    for j in range(1, i + 1):
        print('*', end='')  # Print star without newline
    
    # After inner loop, print a newline to move to the next row
    print()
```

Pattern 2
1
121
12321
1234321

```python
# Take input for number of rows
rows = int(input('Enter number of rows: '))

# Outer loop for each row
for i in range(1, rows + 1):
    
    # Print increasing numbers from 1 to i
    for j in range(1, i + 1):
        print(j, end='')  # Print numbers without newline

    # Print decreasing numbers from i-1 down to 1
    for k in range(i - 1, 0, -1):
        print(k, end='')  # Print numbers without newline

    # Move to next line after finishing a row
    print()
```

Loop Control Statements in Python

Break Statement

Method/Statement: break

Function: Immediately exits the current loop.

Usage: Stop the loop when a certain condition is met.

Continue Statement

Method/Statement: continue

Function: Skips the rest of the current iteration and continues with the next iteration of the loop.

Usage: Ignore specific cases but keep looping.

Pass Statement

Method/Statement: pass

Function: Does nothing; acts as a placeholder.

Usage: Used where a statement is syntactically required but no action is needed.

```python
for i in range(1, 10):
    if i == 5:
        break  # Exit the loop immediately when i is 5
    print(i)    # Print i if loop hasn't been broken
```

```python
# Take input for lower and upper range
lower = int(input('Enter lower range: '))
upper = int(input('Enter upper range: '))

# Loop through each number in the range
for i in range(lower, upper + 1):
    
    # Check if the current number i is divisible by any number from 2 to i-1
    for j in range(2, i):
        if i % j == 0:
            break  # Not a prime number, exit inner loop
    else:
        # This else belongs to the for-loop
        # It runs only if the loop was NOT broken (i.e., i is prime)
        if i > 1:  # 1 is not a prime
            print(i)  # Print the prime number
```

```python
for i in range(1, 10):
    if i == 5:
        continue  # Skip the rest of this iteration when i is 5
    print(i)       # Print i only if it wasn't skipped
```

```python
for i in range(1, 10):
    pass  # Does nothing, acts as a placeholder
```
