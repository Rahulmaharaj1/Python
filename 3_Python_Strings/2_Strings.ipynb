Strings are sequence of Characters
In Python specifically, strings are a sequence of Unicode Character

Strings: Strings are sequences of Unicode characters in Python.

Creating Strings: Strings can be created using single, double, or triple quotes.

Accessing Strings: Access characters using indexing or slicing.

Adding Chars to Strings: Use concatenation (+) or repetition (*).

Editing Strings: Strings are immutable; create a new string to "edit" content.

Deleting Strings: Use del to delete an entire string variable.

Operations on Strings: Perform operations like concatenation, repetition, membership (in), and iteration.

String Functions: Python provides built-in string methods like .upper(), .lower(), .replace(), .split(), .strip().

Creating Stings

```python
# Single quotes
s = 'hello'

# Double quotes
s = "hello"

# Triple quotes for multiline strings
s = '''hello'''
s = """hello"""

# Using the str() constructor
s = str('hello')

# Print the string
print(s)
```

Accessing Substrings from a String

```python
# Positive Indexing

s = 'hello world'
print(s[41])
```

```python
s = 'hello world'

# Negative indexing starts from -1 at the last character
print(s[-3])  # Access the third character from the end
```

```python
s = 'hello world'

# Slice: s[start:end:step]
print(s[6:0:-2])
```

```python
s = 'hello world'

# Reverse the string using slicing
print(s[::-1])
```

```python
s = 'hello world'

# Slice with negative step
print(s[-1:-6:-1])
```

Editing and Deleting in Strings

```python
s = 'hello world'

# Trying to change the first character
s[0] = 'H'  # ❌ This will cause an error
```

```python
s = 'hello world'

# Delete the entire string variable
del s

# Trying to access s now will cause an error
print(s)  # ❌ NameError: name 's' is not defined
```

```python
s = 'hello world'

# Trying to delete a slice from the string
del s[-1:-5:2]  # ❌ Error
print(s)
```
